```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('vgsales.csv')

print("Setir ve sutun sayi:", df.shape)
print("\nSutunlarin veri tipleri:")
print(df.dtypes)
print("\nItkin (bos) deyerlerin sayi:")
print(df.isnull().sum())

df.head()
```

```python
print("Temizlemeden evvel setir sayi:", len(df))

df = df.dropna(subset=['Year'])

df['Publisher'] = df['Publisher'].fillna('Namelum')

print("Temizlemeden sonra setir sayi:", len(df))
print("\nItkin deyerlerin yeni vezyeti:")
print(df.isnull().sum())

df['Year'] = df['Year'].astype(int)

df.head()
```

```python
top10_janr = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(10)
print("En Yaxsi 10 Janr (Qlobal Satis):")
print(top10_janr)

top10_platforma = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
print("\nEn Yaxsi 10 Platforma (Qlobal Satis):")
print(top10_platforma)

top10_nasir = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
print("\nEn Yaxsi 10 Nasir (Qlobal Satis):")
print(top10_nasir)
```

```python
umumi_na = df['NA_Sales'].sum()
umumi_eu = df['EU_Sales'].sum()
umumi_jp = df['JP_Sales'].sum()
umumi_qlobal = df['Global_Sales'].sum()

na_faiz = round(umumi_na / umumi_qlobal * 100, 2)
eu_faiz = round(umumi_eu / umumi_qlobal * 100, 2)
jp_faiz = round(umumi_jp / umumi_qlobal * 100, 2)

print("Simali Amerika (NA) umumi satis:", round(umumi_na, 2), "- Qlobal payi:", na_faiz, "%")
print("Avropa Birliyi (EU) umumi satis:", round(umumi_eu, 2), "- Qlobal payi:", eu_faiz, "%")
print("Yaponiya (JP) umumi satis:", round(umumi_jp, 2), "- Qlobal payi:", jp_faiz, "%")
print("Umumi Qlobal Satis:", round(umumi_qlobal, 2))
```

```python
na_janr = df.groupby('Genre')['NA_Sales'].sum().sort_values(ascending=False)
eu_janr = df.groupby('Genre')['EU_Sales'].sum().sort_values(ascending=False)
jp_janr = df.groupby('Genre')['JP_Sales'].sum().sort_values(ascending=False)

print("Simali Amerikada En Populyar Janr:", na_janr.index[0], "-", round(na_janr.iloc[0], 2))
print("Avropa Birliyinde En Populyar Janr:", eu_janr.index[0], "-", round(eu_janr.iloc[0], 2))
print("Yaponiyada En Populyar Janr:", jp_janr.index[0], "-", round(jp_janr.iloc[0], 2))

regional_janr = pd.DataFrame({
    'NA': na_janr,
    'EU': eu_janr,
    'JP': jp_janr
})

print("\nButun Janrlar uzre Regional Muqayise:")
print(regional_janr)
```

```python
illik_satis = df.groupby('Year')['Global_Sales'].sum()

print("Illik Qlobal Satis:")
print(illik_satis)

print("\nEn yuksek satisli il:", illik_satis.idxmax(), "-", round(illik_satis.max(), 2), "milyon")

plt.figure(figsize=(14, 6))
illik_satis.plot(kind='line', marker='o', color='darkblue', linewidth=2)
plt.title('1980-2016 Illik Qlobal Video Oyun Satis Trendi')
plt.xlabel('Il')
plt.ylabel('Qlobal Satis (milyon nusxe)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
top10_oyun = df.nlargest(10, 'Global_Sales')[['Name', 'Platform', 'Genre', 'Global_Sales']]

print("Butun Zamanlarin En Cox Satilan 10 Oyunu:")
print(top10_oyun.to_string(index=False))
```

```python
def onillik_tap(il):
    if il < 1990:
        return '80-ci illər'
    elif il < 2000:
        return '90-cı illər'
    elif il < 2010:
        return '2000-ci illər'
    else:
        return '2010-cu illər'

df['Onillik'] = df['Year'].apply(onillik_tap)

onillik_orta_satis = df.groupby('Onillik')['Global_Sales'].mean().round(3)

print("Onillik uzre Oyun Basina Orta Satis (milyon nusxe):")
print(onillik_orta_satis.sort_values(ascending=False))

print("\nEn yuksek orta satisli onillik:", onillik_orta_satis.idxmax())
```

```python
en_yaxsi_5_janr = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(5).index.tolist()

qrafik_data = regional_janr.loc[en_yaxsi_5_janr]

qrafik_data.plot(kind='bar', figsize=(12, 6), color=['steelblue', 'coral', 'seagreen'])
plt.title('Simali Amerika, Avropa Birliyi ve Yaponiyada En Yaxsi 5 Janr')
plt.xlabel('Janr')
plt.ylabel('Satis (milyon nusxe)')
plt.xticks(rotation=0)
plt.legend(title='Bolge')
plt.tight_layout()
plt.show()
```

```python
en_yaxsi_10_janr_ustuste = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(10).index.tolist()

qrafik_data2 = regional_janr.loc[en_yaxsi_10_janr_ustuste]

qrafik_data2.plot(kind='bar', stacked=True, figsize=(12, 6), color=['steelblue', 'coral', 'seagreen'])
plt.title('Janr uzre Regional Satislar (Ust-uste Yigilmis)')
plt.xlabel('Janr')
plt.ylabel('Satis (milyon nusxe)')
plt.xticks(rotation=45)
plt.legend(title='Bolge')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top10_nasir.sort_values().plot(kind='barh', color='mediumpurple')
plt.title('En Yaxsi 10 Nasir (Qlobal Satis)')
plt.xlabel('Qlobal Satis (milyon nusxe)')
plt.ylabel('Nasir')
plt.tight_layout()
plt.show()
```
